```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
def isocontour(mus, sigmas, scale):
    x = np.linspace(-scale, scale, 300)
    y = np.linspace(-scale, scale, 300)
    X, Y = np.meshgrid(x, y)
    pos = np.dstack((X, Y))

    
    if len(mus) == 2:
        pdf1 = multivariate_normal(mus[0], sigmas[0])
        pdf2 = multivariate_normal(mus[1], sigmas[1])
        z = pdf1.pdf(pos) - pdf2.pdf(pos)
    else:
        pdf = multivariate_normal(mus[0], sigmas[0])
        z = pdf.pdf(pos)
        
    plt.figure(figsize=(8, 6))

    plt.contourf(X, Y, z)
    plt.colorbar(label='Density')
    plt.xlabel('X')
    plt.ylabel('Y')

    for i in range(len(mus)):
        x = mus[i][0]
        y = mus[i][1]
        plt.scatter(x, y, color="red", s=20)
        plt.annotate(f"Distribution {i}", (x, y), textcoords="offset points", xytext=(0, 10), ha='center', color="red")

    plt.grid()
    plt.show()
```

```python
isocontour([np.array([1, 1])], [np.array([[1, 0], [0, 2]])], 6)
```

```python
isocontour([np.array([-1, 2])], [np.array([[2, 1],[1, 4]])], 6)
```

```python
isocontour([np.array([0, 2]), np.array([2, 0])], [np.array([[2, 1],[1, 1]]), np.array([[2, 1],[1, 1]])], 4)
```

```python
isocontour([np.array([0, 2]), np.array([2, 0])], [np.array([[2, 1],[1, 1]]), np.array([[2, 1],[1, 4]])], 4)
```

```python
isocontour([np.array([1, 1]), np.array([-1, -1])], [np.array([[2, 0],[0, 1]]), np.array([[2, 1],[1, 2]])], 3)
```
