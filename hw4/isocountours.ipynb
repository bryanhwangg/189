```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_lp_norm_contour(p, scale=1.5, resolution=400):
    x = np.linspace(-scale, scale, resolution)
    y = np.linspace(-scale, scale, resolution)
    X, Y = np.meshgrid(x, y)
    
    Lp = (np.abs(X) ** p + np.abs(Y) ** p) ** (1 / p)
    
    plt.figure(figsize=(6, 6))
    plt.contour(X, Y, Lp)
    plt.title(f"L sub {p} norm Contour")
    plt.xlabel('x')
    plt.ylabel('y')
```

```python

plot_lp_norm_contour(p=0.5)
```

```python
plot_lp_norm_contour(p=1)
```

```python
plot_lp_norm_contour(p=2)
```
