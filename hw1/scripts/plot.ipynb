```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
from sklearn.svm import SVC 
```

```python
data = np.load("../data/toy-data.npz")
data.files
```

```python
training_data = data['training_data']
training_labels = data['training_labels']
training_data.shape, training_labels.shape
```

```python
x = training_data[:, 0]
y = training_data[:, 1]
w = np.array([-0.4528, -0.5190])
a = 0.1471
```

```python
# Plot the data points
def plot_data_points(data, labels):
    plt.scatter(data[:, 0], data[:, 1], c=labels)
# Plot the decision boundary
def plot_decision_boundary(w, b):
    x = np.linspace(-5, 5, 100)
    y = -(w[0] * x + b) / w[1]
    plt.plot(x, y)
# Plot the margins
def plot_margins(w, b):
    norm = sum(np.sqrt(w**2))
    x = np.linspace(-5, 5, 100)
    y_pos = -(w[0] * x + b + norm) / w[1]
    y_neg = -(w[0] * x + b - norm) / w[1]
    plt.plot(x, y_pos, ls="-", label="+ margin") 
    plt.plot(x, y_neg, ls="-", label="- margin")
    plt.legend()
def plot_support_vectors():
    clf = svm.SVC(kernel="linear")
    clf.fit(training_data, training_labels)
    support_vectors = clf.support_vectors_
    x = support_vectors[:, 0]
    y = support_vectors[:, 1]
    plt.scatter(x, y, color="blue")
```

```python
plot_data_points(training_data, training_labels)
plot_decision_boundary(w, 0.1471)
plot_margins(w, a)
plot_support_vectors()
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("SVM on toy data (blue are support vectors)")
```
